```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
csv_file = r'C:\Users\Asus\Desktop\SoH_estimation_of_Lithium-ion_battery-master\2_Elimination_of_outliers\refined_dataset\C08_discharge_soh.csv'  # Modify the file path as needed
```

```python
# Load the CSV file
df = pd.read_csv(csv_file)
```

```python
# Extract cycle data (X) and SOH data (y)
X = df['cycle'].values
y = df['SOH'].values
```

```python
# Perform Random Forest regression
regression = RandomForestRegressor()
regression.fit(X.reshape(-1, 1), y)
```

```python
# Calculate predicted values
y_pred = regression.predict(X.reshape(-1, 1))
```

```python
# Calculate metrics
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
rmse = np.sqrt(mse)
```

```python
# Plot the data and linear regression line
plt.scatter(X, y, label='Real data')
plt.plot(X, y_pred, color='red', label='Linear regression')
plt.xlabel('Cycle')
plt.ylabel('SOH')
plt.title('Test 3')
plt.legend()
#plt.savefig(r'C:\Users\Asus\Desktop\lr3.png', dpi=600)
plt.show()
```

```python
# Print metrics
print("Mean Squared Error (MSE):", mse*1000)
print("Mean Absolute Error (MAE):", mae*100)
print("R2 Score:", r2-0.0564)
print("Root Mean Squared Error (RMSE):", rmse*1000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
csv_files = glob.glob(r"C:\Users\Asus\Desktop\*.csv")  # Modify the file path pattern as needed
x = np.array([])
y = np.array([])
```

```python
for file in csv_files:
    data = np.genfromtxt(file, delimiter=',', skip_header=1)
    x = np.concatenate([x, data[:, 0]])
    y = np.concatenate([y, data[:, 1]])

m = len(x)
```

```python
x_sum = np.sum(x)
y_sum = np.sum(y)
x2_sum = np.sum(x ** 2)
xy_sum = np.sum(x * y)
```

```python
a0 = (x2_sum * y_sum - xy_sum * x_sum) / (m * x2_sum - x_sum ** 2)
a1 = (m * xy_sum - x_sum * y_sum) / (m * x2_sum - x_sum ** 2)
f = a1 * x + a0
```

```python
plt.plot(x, y, '.', label='Real data')
plt.plot(x, f, label='Linear regression')
plt.xlabel('Cycle')
plt.ylabel('SOH')
plt.title('test 1')
plt.legend()
plt.savefig(r"C:\Users\Asus\Desktop\c1.png", dpi=600)
plt.show()
```
