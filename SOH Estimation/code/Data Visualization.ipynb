```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
csv_files = glob.glob(r"C:\Users\Asus\Desktop\SoH_estimation_of_Lithium-ion_battery-master\2_Elimination_of_outliers\refined_dataset\*.csv")  # Modify the file path pattern as needed

data_frames = []
```

```python
# Load and concatenate data frames
for file in csv_files:
    df = pd.read_csv(file)
    data_frames.append(df)

merged_df = pd.concat(data_frames, ignore_index=True)
```

```python
# Select relevant columns
columns_to_plot = ['terminal_voltage', 'terminal_current', 'temperature', 'capacity', 'cycle', 'SOH']
selected_df = merged_df[columns_to_plot]
```

```python
# Compute correlation matrix
corr_matrix = selected_df.corr()
```

```python
# Generate heat map
plt.figure(figsize=(12, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heat Map')
plt.savefig(r"C:\Users\Asus\Desktop\heatmap.png", dpi=600)
plt.show()
```

# diag_kind options:

'auto' (default): Automatically select the appropriate plot type based on the data type of the variable.

'hist': Plot a histogram on the diagonal for numeric variables.

'kde': Plot a kernel density estimate on the diagonal for numeric variables.

'scatter': Plot a scatter plot on the diagonal for numeric variables.

'reg': Plot a regression line on the diagonal for numeric variables.

'none': Do not plot anything on the diagonal.

# plot_kws options:

dict: A dictionary of keyword arguments that are passed to the underlying plotting function. For example, plot_kws={'color': 'red', 'marker': 'o'} would set the color of the dots to red and use circular markers.

Any additional valid arguments specific to the underlying plotting function can also be used. For example, for scatter plots, you can use plot_kws={'s': 10} to set the size of the dots.

```python
# Select relevant columns
columns_to_plot = ['temperature', 'capacity', 'cycle', 'SOH']
selected_df = merged_df[columns_to_plot]
```

```python
# Compute correlation matrix
corr_matrix = selected_df.corr()
```

```python
# Generate pair plot
sns.pairplot(selected_df, diag_kind='kde', plot_kws={'s': 4})
plt.savefig(r"C:\Users\Asus\Desktop\pairplot1.png", dpi=600)
plt.show()
```
